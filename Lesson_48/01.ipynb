Застосувати методи трансформації даних до датасету `Data_to_Transform`.

## Рішення

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Data_to_Transform.csv')
df.head()
```

```python
fig, ax = plt.subplots(1,4, figsize=(17, 5))
for i in range(len(df.columns)):
    sns.distplot(df[df.columns[i]], ax=ax[i])
```

```python
fig, ax = plt.subplots(1,4, figsize=(17, 5))
for i in range(len(df.columns)):
    stats.probplot(df[df.columns[i]], plot=ax[i])
```

### Кубічний корінь

```python
fig, ax = plt.subplots(1,4, figsize=(17, 5))
for i in range(len(df.columns)):
    sns.distplot(np.cbrt(df[df.columns[i]]), ax=ax[i])
```

```python
fig, ax = plt.subplots(1,4, figsize=(17, 5))
for i in range(len(df.columns)):
    stats.probplot(np.cbrt(df[df.columns[i]]), plot=ax[i])
```

### Логарифмування

```python
fig, ax = plt.subplots(1,4, figsize=(17, 5))
for i in range(len(df.columns)):
    sns.distplot(np.log(df[df.columns[i]]), ax=ax[i])
```

```python
fig, ax = plt.subplots(1,4, figsize=(17, 5))
for i in range(len(df.columns)):
    stats.probplot(np.log(df[df.columns[i]]), plot=ax[i])
```

### Перетворення Бокса-Кокса

```python
fig, ax = plt.subplots(1,4, figsize=(17, 5))
for i in range(len(df.columns)):
    sns.distplot(stats.yeojohnson(df[df.columns[i]])[0], ax=ax[i])
```

```python
fig, ax = plt.subplots(1,4, figsize=(17, 5))
for i in range(len(df.columns)):
    stats.probplot(stats.yeojohnson(df[df.columns[i]])[0], plot=ax[i])
```
