# Завдання 1

Два економісти заповнюють документи з власними прогнозами щодо курсу певних цінних паперів, після чого їх складають у спільну папку. Ймовірність відхилення
прогнозованого значення курсу від фактичного понад допустиму похибку для першого економіста дорівнює 0,1; для другого 0,2. Перший економіст заповнив 40 документів другий - 60. Під час перевірки навмання взятий із папки документ виявився з суттєвою похибкою. Визначити ймовірність того, що його склав перший економіст.

## Рішення:
- $H_1$ - документ заповнив перший економіст;
- $H_2$ - документ заповнив другий економіст;
- $A$ - взятий документ з похибкою;

Тоді:
- $P(H_1) = \frac{40}{100} = 0.4$
- $P(H_2) = \frac{60}{100} = 0.6$
- $P(A|H_1) = 0.1$
- $P(A|H_2) = 0.2$
- $P(A) = P(A|H_1) \times P(H_1) + P(A|H_1) \times P(H_2) = 0.1 \times 0.4 + 0.2 \times 0.6 = 0.16$

Використовуємо формулу Байєса:
$P(H_1|A) = \frac{P(H_1)P(A|H_1)}{P(A)} = \frac{0.4 \times 0.1}{0.16} = 0.25$

**Відповідь:** ймовірність того, що документ з похибкою був складений першим економістом складає 25%.
