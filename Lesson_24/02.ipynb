Є можливість вибрати спосіб виробництва і реалізації двох наборів товарів широкого вжитку. За даними відділу маркетингу, яким були проведені дослідження ринку, можливий прибуток від виробництва і реалізації $Х$ і $У$ наведено в таблицях ($X$, $Y$ - прибуток у грошових одиницях):

| **X** | 1000 | 1500 | 2000 |
|-------|------|------|------|
| **P** | 0.5  | 0.3  | 0.2  |

| **Y** | 1000 | 1500 | 1750 |
|-------|------|------|------|
| **P** | 0.4  | 0.4  | 0.2  |

Потрібно оцінити ступінь ризику і запропонувати рішення щодо випуску і реалізації одного із наборів товарів.

## Рішення:
Використаємо масиви `numpy` для вирішення задачі:

```python
import numpy as np

X = np.array([1000, 1500, 2000])
Px = np.array([0.5, 0.3, 0.2])

# Математичне сподівання:
Mx = sum(X * Px)

# Стандартне відхилення:
SDx = (sum(X**2 * Px) - Mx ** 2) ** 0.5
print(f'Математичне сподівання прибутку від товару X: {Mx}\n'
      f'Стандартне відхилення прибутку від товару X: {round(SDx, 2)}')
```

```python
Y = np.array([1000, 1500, 1750])
Py = np.array([0.4, 0.4, 0.2])

# Математичне сподівання:
My = sum(Y * Py)

# Стандартне відхилення:
SDy = (sum(Y**2 * Py) - My ** 2) ** 0.5
print(f'Математичне сподівання прибутку від товару Y: {My}\n'
      f'Стандартне відхилення прибутку від товару Y: {round(SDy, 2)}')
```

**Відповідь:** Математичне сподівання прибутку від обох товарів однакова. Менш ризикованим буде випуск товару $Y$, оскільки стандартне відхилення прибутку у нього менше.
