# Завдання
Створити власний датасет на основі випадкових даних за допомогою `np.linspace` та побудуй на них лінійну регресію використовуючи `tensorflow`.

## Рішення

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# Створення випадкових даних
x = np.linspace(0, 10, 100)
y = 2 * x + np.random.randn(100) * 2

# Побудова моделі лінійної регресії використовуючи TensorFlow
inputs = tf.keras.layers.Input(shape=(1,))
hidden1 = tf.keras.layers.Dense(64, activation='relu')(inputs)
outputs = tf.keras.layers.Dense(1)(hidden1)

model = tf.keras.models.Model(inputs=inputs, outputs=outputs)

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

history = model.fit(x, y, epochs=200, verbose=0)

# Виведення графіку
y_pred = model.predict(x)

plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.show()
```
