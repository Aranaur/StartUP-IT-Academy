При вивченні випадкової величини $Х$ у результаті 40 незалежних спостережень дістали вибірку: 10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 14, 13, 9, 8, 8, 7, 10, 10, 11, 11, 11, 12, 8, 7,9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12. Побудувати дискретний статистичний розподіл для цієї вибірки, а також полігон частот і $F(x)$. Знайти моду та медіану.

## Рішення:
1. Завантажуємо модулі та формуємо розподіл дискретної величини:

```python
from collections import Counter
import statistics
from tomark import Tomark
from IPython.display import Markdown as md

X = [10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 14, 13, 9, 8, 8, 7,
    10, 10, 11, 11, 11, 12, 8, 7, 9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12]

X_count = {k: v for k, v in sorted(Counter(X).items(), key=lambda item: item[0])}
X_freq = {k: v / len(X) for k, v in sorted(Counter(X).items(), key=lambda item: item[0])}

data = [
    X_count,
    X_freq
]

markdown = Tomark.table(data)
md(markdown)
# X_table = {}
# for k in set(X_count) & set(X_freq):
#     X_table[k] = [X_count[k], X_freq[k]]
# print(X_table)
```

2. Розрахунок моди та медіани:

```python
print(f'Медіана дискретної випадкової величини дорівнює {statistics.median(X)}')
print(f'Мода дискретної випадкової величини дорівнює {statistics.mode(X)}')
```

3. Полігон частот

```python
import matplotlib.pyplot as plt
x = X_count.keys()
y = X_count.values()
plt.plot(x, y, '-')
plt.plot(x, y, '.')
plt.xlabel('X')
plt.ylabel('Частота')
plt.show()
```

4. Полігон відносних частот:

```python
x = X_freq.keys()
y = X_freq.values()
plt.plot(x, y, '-')
plt.plot(x, y, '.')
plt.xlabel('X')
plt.ylabel('Відносна Частота')
plt.show()
```
