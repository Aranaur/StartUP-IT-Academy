# ЗАВДАННЯ 1
Прилад містить 7 елементів, з яких три зношені. При включенні приладу випадковим чином включається три елементи. Знайти ймовірність того, що при цьому виявляється включеними два незношені елементи.

## Рішення
**ВАРІАНТ 1:**
Закодуємо зношені (0) та незношені елементи (1).
Припустимо, що елементи вмикаються послідовно.

Існує три варіанти роботи приладу за умови двох незношених та одного зношеного елементу: **110**, **101**, **011**.

Тоді ймовірність таких подій буде $\frac{4}{7} \times  \frac{3}{6} \times  \frac{3}{5} = \frac{36}{210} = \frac{6}{35} \approx 0.17$

Загальна ймовірність будь-якого з тьох випадків: **$3 * \frac{6}{35} = \frac{18}{35} \approx 0.51$**

**Відповідь: 51%**

**ВАРІАНТ 2:**
Загальна формула підрахунку кількості варіантів:
$$C_{m}^{n} = \frac{n!}{m!(n-m)!}$$

1. Визначимо кількість подій, які можуть відбутися - включаться три елементи з 7: $n = C_{7}^{3} = \frac{7!}{3!4!} = 35$
2. Дві браковані деталі можуть включитися у наступній кількості подій: $C_{4}^{2}$
3. Одна ціла деталь може бути включена у наступній кількості подій: $C_{3}^{1}$
4. Дві браковані і одна ціла: $C_{4}^{2} \times C_{3}^{1}$
5. Тоді ймовірність описаної події: $\frac{C_{4}^{2} \times C_{3}^{1}}{35} = \frac{\frac{4!}{2!*2!} \times \frac{3!}{1!\times2!}}{35} = \frac{18}{35} \approx 0.51$

**Відповідь: 51%**
