# ЗАВДАННЯ 2
Для сигналізації про аварію встановлено два незалежно працюючих сигналізатори. Ймовірність того, що при аварії спрацює перший сигналізатор дорівнює $p_1 = 0,95$, ця ймовірність для другого сигналізатора дорівнює $p_1 = 0,9$. Знайти ймовірність того, що при аварії спрацює:

а) хоча б один сигналізатор;

б) тільки один сигналізатор.

### Рішення
Ймовірність того, що перша сигналізація не спрацює:
$q_1 = 1 - 0.95 = 0.05$

Ймовірність того, що друга сигналізація не спрацює:
$q_2 = 1 - 0.9 = 0.1$

а) Ймовірність того, що обидві сигналізації не спрацюють:
$q = q_1q_2 = 0.05*0.1 = 0.005$

Ймовірність того, що спрацює хоча б один сигналізатор:
$p = q - 1 = 1 - 0.005 = 0.995$

**Відповідь: 99,5%**

б) Ймовірність того, що спрацює тільки один сигналізатор (теорема додавання та множення несумісних подій):
$p_1 = p_1\times q_2 + q_1\times p_2 = 0.95\times 0.1 + 0.05\times 0.9 = 0.14$

**Відповідь: 14%**
