# ЗАВДАННЯ 3
Два верстати-автомати виготовляють однакові вироби, які попадають на загальний конвеєр. Продуктивність першого автомата в два раза більш ніж продуктивність другого. Перший верстат-автомат виготовляє $P(A) = 0.6$ виробів вищого гатунку, а другий - $P(B) = 0.84$.

а) Знайти ймовірність того, що навмання взятий з конвеєра виріб вищого ґатунку.

б) Навмання взятий з конвеєра виріб має вищий ґатунок. Знайти ймовірність того, що він виготовлений на першому верстаті.

### Рішення
а) Ймовірність тощо, що виріб зроблено на першому автоматі $\frac{2}{3}$, а на другому відповідно $\frac{1}{3}$

**Ймовірність навмання вибраного виробу вищого гатунку:**

$P(C) = P(A) \times \frac{2}{3} + P(B) \times  \frac{1}{3} = 0.6\times \frac{2}{3} + 0.84 \times  \frac{1}{3} = 0.68$

**Відповідь: 68%**

б) Ймовірність того, що виріб виготовлений на першому верстаті:

$P(D) = \frac{\frac{2}{3} \times P(A)}{P(C)} = \frac{\frac{2}{3} \times 0.6}{0.68} = 0.588$

**Відповідь: 58,8%**
