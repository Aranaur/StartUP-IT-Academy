## Рішення

```python
import cv2
import matplotlib.pyplot as plt
import dlib
from imutils import face_utils
```

```python
# Завантажуємо зображення та класифікатор
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
img = cv2.imread("titanic.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(img_gray, 1.3, 5)
for face in faces:
    x, y, w, h = face
    cv2.rectangle(img_rgb, (x, y), (x+w, y+h), (255, 0, 0), 5)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
img_90 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
img_gray = cv2.cvtColor(img_90, cv2.COLOR_BGR2GRAY)
img_90_rgb = cv2.cvtColor(img_90, cv2.COLOR_BGR2RGB)
faces = face_cascade.detectMultiScale(img_gray, 1.3, 5)
for face in faces:
    x, y, w, h = face
    cv2.rectangle(img_90_rgb, (x, y), (x+w, y+h), (255, 0, 0), 5)
plt.imshow(img_90_rgb)
plt.axis('off')
plt.show()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
face_detect = dlib.get_frontal_face_detector()
rects = face_detect(img_rgb, 1)

for (i, rect) in enumerate(rects):
    (x, y, w, h) = face_utils.rect_to_bb(rect)
    cv2.rectangle(img_rgb, (x, y), (x + w, y + h), (255, 255, 255), 3)
    
plt.figure(figsize=(10,6))
plt.imshow(img_rgb, cmap='gray')
plt.axis('off')
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
face_detect = dlib.get_frontal_face_detector()
rects = face_detect(img_gray, 1)

for rect in rects:
    (x, y, w, h) = face_utils.rect_to_bb(rect)
    cv2.rectangle(img_gray, (x, y), (x + w, y + h), (255, 255, 255), 3)
    
plt.figure(figsize=(10,6))
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.show()
```
